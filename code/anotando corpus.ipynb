```python
combinations = ["only-tokens","fusion","combination","annotation"]
```

```python

for corpus in corpora_lt:
    print(corpus)
    # where the TEI master files are
    infolder = "/home/jose/Dropbox/MTB/investigacion/estudio_parametros/corpora/" + corpus
    
    for combination in combinations:
        print(infolder + "/" + combination)
        if os.path.exists(infolder + "/" +  combination):
            shutil.rmtree(infolder + "/" +  combination)
        os.makedirs(infolder + "/" + combination)

    for doc in glob.glob(infolder + "/annotated-xml/*.xml"):
        root = etree.parse(doc).getroot()

        input_name  = os.path.splitext(os.path.split(doc)[1])[0]
        print(input_name)

        for combination in combinations:
            content = """"""
            if combination == "only-tokens":
                for w in root.xpath('//w[@form][@pos!="punctuation"]', with_tail=True):
                    content += "\n" + w.xpath('./@form', with_tail=True)[0]
            elif combination == "fusion":
                for w in root.xpath('//w[@form][@pos!="punctuation"]', with_tail=True):
                    content += "\n" +  w.xpath('./@form', with_tail=True)[0] +"_" +  w.xpath('./@ctag', with_tail=True)[0]

            elif combination == "combination":
                for w in root.xpath('//w[@form][@pos!="punctuation"]', with_tail=True):
                    content += "\n" + w.xpath('./@form', with_tail=True)[0] +" " +  w.xpath('./@ctag', with_tail=True)[0]

            elif combination == "annotation":
                for w in root.xpath('//w[@form][@pos!="punctuation"]', with_tail=True):
                    content += "\n" + w.xpath('./@tag')[0] + " " +   w.xpath('./@ctag')[0] 

            with open (infolder + "/" + combination + "/"+ input_name +".txt", "w", encoding="utf-8") as fout:
                fout.write(content)
    print("\n\n")
```
